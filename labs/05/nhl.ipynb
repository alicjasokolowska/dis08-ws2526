```python
import pandas as pd

# von Chrome add on gecraped-te csv-Datei laden
df = pd.read_csv("data.csv")  

# year muss numerisch sein
df["year"] = pd.to_numeric(df["year"], errors="coerce")
```

```python
print("Who made the most wins in 1990, 2000, and 2010? ")
for year in [1990, 2000, 2010]:
    subset = df[df["year"] == year]
    top = subset.sort_values("wins", ascending=False).iloc[0]
    print(f"{year}: {top['name']} mit {top['wins']} Wins")
```

```python
print("How many teams participated in 1991, 2001, and 2011?")
for year in [1991, 2001, 2011]:
    count = df[df["year"] == year]["name"].nunique()
    print(f"{year}: {count} Teams")
```
