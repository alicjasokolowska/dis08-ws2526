```python
from ucimlrepo import fetch_ucirepo
from scipy import stats
import pandas as pd

# Datensatz laden
adult = fetch_ucirepo(id=2)

# Features & Target
X = adult.data.features
y = adult.data.targets

# Target zu Series machen
income = y.iloc[:, 0]

# Relevante Variable auswählen
hours = X["hours-per-week"]

# Zwei unabhängige Stichproben
hours_le50 = hours[income == "<=50K"].dropna()
hours_gt50 = hours[income == ">50K"].dropna()

# t-Test (Welch)
t_stat, p_value = stats.ttest_ind(
    hours_le50,
    hours_gt50,
    equal_var=False
)

print("Mean hours (<=50K):", hours_le50.mean())
print("Mean hours (>50K):", hours_gt50.mean())
print("t-statistic:", t_stat)
print("p-value:", p_value)
```

```python
# Der durchgeführte T-Test ergibt einen p-Wert kleiner als 0,05. Das bedeutet, dass der beobachtete Unterschied in den durchschnittlichen Arbeitsstunden nicht zufällig ist.
# Daher wird die Nullhypothese verworfen und die Alternativhypothese angenommen.
# Es zeigt sich, dass Personen mit einem Einkommen von über 50.000 Dollar im Durchschnitt mehr Stunden pro Woche arbeiten als Personen mit einem Einkommen von 50.000 Dollarr oder weniger.
```
