```python
def commaCode(items):
    if len(items) == 0:
        return ''
    elif len(items) == 1:
        return items[0]
    else:
        return ', '.join(items[:-1]) + ', und ' + items[-1]


spam = ['traube', 'apfel', 'birne', 'zitrone']
empty_list = []
single_item = ['tomate']

print(commaCode(spam))        # traube, apfel, birne, zitrone
print(commaCode(empty_list))  # leere Liste
print(commaCode(single_item)) # tomate
```

```python
import random

numberOfStreaks = 0

for experimentNumber in range(10000):
    flips = []
    for i in range(100):
        flips.append('H' if random.randint(0, 1) == 0 else 'T')
    
    # 6-er Streak 
    for i in range(len(flips) - 5):
        if flips[i] == flips[i+1] == flips[i+2] == flips[i+3] == flips[i+4] == flips[i+5]:
            numberOfStreaks += 1
            break  # nur einmal pro 100 Würfe zählen

print('Chance of streak: %s%%' % (numberOfStreaks / 100))
```

```python
stuff = {'rope': 1, 'torch': 6, 'gold coin': 42, 'dagger': 1, 'arrow': 12}

def displayInventory(inventory):
    print("Inventory:")
    item_total = 0
    for k, v in inventory.items():
        print(f"{v} {k}")
        item_total += v
    print("Total number of items: " + str(item_total))


displayInventory(stuff)
```

```python
def addToInventory(inventory, addedItems):
    for item in addedItems:
        inventory[item] = inventory.get(item, 0) + 1
    return inventory


inv = {'gold coin': 42, 'rope': 1}
dragonLoot = ['gold coin', 'dagger', 'gold coin', 'gold coin', 'ruby']
inv = addToInventory(inv, dragonLoot)
displayInventory(inv)
```
